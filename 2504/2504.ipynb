#Introdução ao Aprendizado de Maquina

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()
df['celsius'] = np.arange(-10, 40)
df.head()
```

```python
def Celsius2fahrenheit(celsius):
  return celsius * 9/5 + 32
```

```python
df['fahrenheit'] = Celsius2fahrenheit(df['celsius'])
df.head(11)
```

## Regressão linear
###Se tivermos um conjunto de dados com valores de temperaturas em  oC  (Celsius) e seus respectivos valores em  oF  (Fahrenheit), podemos criar um programa que aprende esse padrão e empregá-lo para a conversão de novos dados.

```python
from sklearn.linear_model import LinearRegression

X = df[['celsius']] 
y = df['fahrenheit']

regressor = LinearRegression()
regressor.fit(X, y)
```

```python
Fahrenheit_prediction = regressor.predict(pd.DataFrame({'celsius':[0,27]}))
print(Fahrenheit_prediction)
```

##Alterando o programa anterior

```python
from sklearn.linear_model import LinearRegression

X = df[['fahrenheit']]
y = df['celsius'] 

regressor = LinearRegression()
regressor.fit(X, y)
```

```python
Celsius_prediction = regressor.predict(pd.DataFrame({'fahrenheit':[32,80.6]}))
print(Celsius_prediction)
```

##Outro dado que pode ser inserido

```python
df['kelvin'] = (df['fahrenheit'] - 32) * 5/9 + 273.15
df.head(11)
```

```python
from sklearn.linear_model import LinearRegression

X = df[['celsius']]
y = df['kelvin'] 

regressor = LinearRegression()
regressor.fit(X, y)

Kelvin_prediction = regressor.predict(pd.DataFrame({'celsius':[27,0]}))
print(Kelvin_prediction)
```

###Sobre o Aprendizado de Maquina

O Aprendizado de Máquina se aplica a problemas em que não conhecemos as regras explícitas que mapeam as entradas e saídas desejadas. Se a regra é conhecida o modelo explícito será sempre será mais simples e terá melhores resultados.

##Terminologia

###Rótulos ou Variável Objetivo,  y

É a saída que estamos buscando. Muitas vezes assume valores discretos, chamados de classes, ou grupos

###Recursos, Features ou Atributos,  X

É uma variavel de entrada, que pode ter uma ou mais dimensões, e corresponde a caracteristicas daquela instancia e que, iamginamos, possam dizer algo sobre o nosso objetivo y

###Conjunto de Treinamento

São dados que apresentam um rotulo, ou variavel objetivo, que desejamos conhecer

###Modelos

Ele que define a relação entre os recrusos e o rotulo, ou seja, entre as entradas e saidas.

##Tipos de Aprendizado de Maquina

###Aprendizado supervisionado

- Classificação
- Regressão

###Aprendizado Não Supervisionado

- Agrupamento
- Associação
- Sumarização

###Apredizado de Reforço

- Movimentação Autonoma
