# Dados

Este é um conjunto de dados de casos de pagamentos inadimplentes (default) de clientes de cartões em Taiwan ([UCI Dataset](https://archive.ics.uci.edu/dataset/350/default+of+credit+card+clients)).

```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/default_credit_cards.csv')
df.head()
```

## Exploração e Preparação dos dados

### Dados ausentes

Elimine dados ausentes se houverem. Verifique os dados categóricos e havendo valores inválidos atribua 'others'.

```python
df.isnull().sum()
```

```python
colunas = df.columns
print(colunas)
```

```python
for column in colunas:
  print(df[column].unique())
```

```python
wrong_columns = ['MARRIAGE', 'EDUCATION']
true_values_marriage = ['married', 'single', 'others']
true_values_education = ['university', 'graduate school', 'high school', 'others']

for i in wrong_columns:
  if i == 'MARRIAGE':
    for value in df[i]:
      if value not in true_values_marriage:
        print('caiu aqui')
        df[i] = df[i].replace(value, 'others')
  elif i == 'EDUCATION':
    for value in df[i]:
      if value not in true_values_education:
        print('caiu aqui')
        df[i] = df[i].replace(value, 'others')


for column in wrong_columns:
  print(df[column].unique())
```

```python
df.MARRIAGE = df.MARRIAGE.replace('0', 'others')
df.EDUCATION = df.EDUCATION.replace(['0', '5', '6'], 'others')

print(df['MARRIAGE'].unique())
print(df['EDUCATION'].unique())
```

### Hot encode

Faça o hot encode dos dados categóricos eliminando a primeira categoria (colinearidade).

Dica: Empregue `hot_encode.get_feature_names_out(['SEX', 'EDUCATION', 'MARRIAGE'])` para nomear os atributos.



```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first')
encoder.fit(df[['EDUCATION', 'MARRIAGE', 'SEX']])
encoder.transform(df[['EDUCATION', 'MARRIAGE', 'SEX']])
```

```python
encoder.categories_
```

```python
encoder.get_feature_names_out(['EDUCATION', 'MARRIAGE', 'SEX'])
```

```python
df_encoder = pd.DataFrame(encoder.transform(df[['EDUCATION', 'MARRIAGE', 'SEX']]), columns=encoder.get_feature_names_out(['EDUCATION', 'MARRIAGE', 'SEX']))
df_encoder
```

Q1. Qual a soma dos hot encodes obtidos?

```python
df_encoder.sum().sum() # df_hot_encode é um dataframe somente com os hot encodes
```

### Normalização

Sugestão: crie um dataframe `df_` concatenando os valores hot encode obtidos e os demais atributos preditores de `df`. Em seguida faça a normalização standard scale dos dados. Ao final, se quiser adicione a coluna da variável objetivo.

```python
new_df_encoder = pd.concat([df, df_encoder], axis=1)
new_df_encoder
```

```python
new_df_encoder.columns
```

```python
new_df_encoder.drop(columns=['ID', 'EDUCATION', 'MARRIAGE', 'SEX', 'default payment next month'], inplace=True)
```

Q2. Qual o shape dos dados obtido?

```python
new_df_encoder.shape
```

#### Standard Scale

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, MaxAbsScaler

scaler = StandardScaler()
scaler.fit(new_df_encoder)
df_scaled = scaler.transform(new_df_encoder)
```

Q3. Qual a soma dos dados do novo dataframe normalizado, incluindo a variável objetivo?

```python
df_scaled.sum().sum() + df['default payment next month'].sum()
```

```python
# Seu código
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split

X = df_scaled
y = df['default payment next month']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=1)

# Definição do Modelo
clf = LogisticRegression(max_iter=1000)
# clf = KNeighborsClassifier(n_neighbors = 5)

# Treinamento
clf.fit(X_train,y_train)

# Predição
y_pred = clf.predict(X_test)

# Métricas
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

cm = confusion_matrix(y_test, y_pred)
print('\nMatriz de Confusão:\n')
print(cm)

accuracy = accuracy_score(y_test, y_pred)
print('\nScore de Acuracidade (1):\n')
print(f'{accuracy:.4f}')

print('\n', clf)

accuracy = clf.score(X_test, y_test)
print('\nScore de Acuracidade (2):\n')
print(f'{accuracy:.4f}')

print('\nClassification Report:\n')
print(classification_report(y_test, y_pred))
```
