```python
mortes_por_alzheimer = df["Deaths - Alzheimer's disease and other dementias - Sex: Both - Age: All Ages (Number)"].sum()
mortes_por_alzheimer
```

```python
mortes_por_alzheimer_AFG = df.loc[df["Entity"]=='Afghanistan']["Deaths - Alzheimer's disease and other dementias - Sex: Both - Age: All Ages (Number)"].sum()
mortes_por_alzheimer_AFG
```

```python
afg_sobre_geral = mortes_por_alzheimer_AFG / mortes_por_alzheimer
afg_sobre_geral_pctg = afg_sobre_geral * 100
print(f'O Afeganistão representa um total de % {afg_sobre_geral_pctg :.3f} em Mortes por Alzheimer no mundo todo.')
```

```python
entities = df['Entity']
entities
```

```python
entidades = entities.unique()
entidades
```

```python
for i in entidades:
  mortes_por_alzheimer_mundo = df.loc[df["Entity"]==i]["Deaths - Alzheimer's disease and other dementias - Sex: Both - Age: All Ages (Number)"].sum()
  pais_sobre_geral = mortes_por_alzheimer_mundo / mortes_por_alzheimer
  pais_sobre_geral_pctg_mundo = pais_sobre_geral * 100
  print(f'O {i} representa um total de % {pais_sobre_geral_pctg_mundo :.3f} em Mortes por Alzheimer no mundo todo.')
```

```python
infos_pctg_mortes_pais = {}
for g in entidades:
  mortes_por_alzheimer_mundo = df.loc[df["Entity"]==g]["Deaths - Alzheimer's disease and other dementias - Sex: Both - Age: All Ages (Number)"].sum()
  pais_sobre_geral = mortes_por_alzheimer_mundo / mortes_por_alzheimer
  pais_sobre_geral_pctg_mundo = pais_sobre_geral * 100
  infos_pctg_mortes_pais[g] = round(pais_sobre_geral_pctg_mundo, 2)

infos_pctg_mortes_pais
```
