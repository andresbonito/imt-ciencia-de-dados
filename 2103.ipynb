## Exercicio 2 - plt.plot(), random data

```python
df = pd.DataFrame()
df['x-values'] = np.arange(0,10,0.1)
df['y-values'] = np.random.sample(len(df))
df
```

```python
plt.plot(df['x-values'], (df['x-values']/df['y-values']), '--r', linewidth=1, markersize=3)
plt.title('x/y')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.show()
```

```python
df.sort_values(by='timepoint')
```

##Exercicio 3

```python
subjectS0 = df[df['subject'] == 's0'].sort_values('timepoint')

region_parietal = subjectS0[subjectS0['region'] == 'parietal']

fmri = region_parietal[region_parietal['event'] == 'stim']

signal = fmri['signal']

timepoint = fmri['timepoint']
subjectS0
```

```python
plt.figure(figsize=(6,5))
plt.plot(timepoint, signal, color='blue', linestyle='solid', linewidth=3, markersize=1)
plt.title('Subjetc S0')
plt.ylabel('Signal')
plt.xlabel('Time')
plt.show()
```

##Exercicio 4

###1Âª soluÃ§Ã£o, mas errada

Foram criadas arias variveis, o que atrapalhou a criaÃ§Ã£o do grafico de forma correta

```python
cue = subjectS0[subjectS0['event'] == 'cue']
signal_cue = cue['signal']
timepoint_cue = cue['timepoint']

stim = subjectS0[subjectS0['event'] == 'stim']
signal_stim = stim['signal']
timepoint_stim = stim['timepoint']
```

```python
plt.figure(figsize=(6,5))
plt.plot(timepoint_stim, signal_stim, color='blue', linestyle='solid', linewidth=3, markersize=1)
plt.plot(timepoint_cue, signal_cue, color='red', linestyle='solid', linewidth=3, markersize=1)
plt.title('Subjetc S0')
plt.ylabel('Signal')
plt.xlabel('Time')
plt.show()
```

###2Âª soluÃ§Ã£o

```python
display(df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'stim')].sort_values('timepoint'))

display(df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'cue')].sort_values('timepoint'))
```

```python
plt.figure(figsize=(12,5))
plt.plot(df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'stim')].sort_values('timepoint').timepoint,
     df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'stim')].sort_values('timepoint').signal,
     color='blue', linestyle='dashdot', linewidth=3, markersize=1)
plt.plot(
    df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'cue')].sort_values('timepoint').timepoint,
    df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'cue')].sort_values('timepoint').signal,
    color='red', linestyle='dashdot', linewidth=3, markersize=1)
plt.legend(['stim', 'cue'])
plt.title('Subjetc S0')
plt.ylabel('Signal')
plt.xlabel('Time')
plt.show()
```
