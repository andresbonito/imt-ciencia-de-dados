```python
chargers = df['charges']
media_chargers = chargers.mean()
print(f'Q3 - A media de pagamentos dos chargers efetuados foi de {media_chargers :.2f}')
```

```python
# Q4

dez_maiores_chargers = chargers.nlargest(10)
dez_maiores_chargers
```

```python
media_dez_maiores_chargers = dez_maiores_chargers.mean()
media_dez_maiores_chargers
```

```python
print(f'Q4 - A media de pagamentos dos 10 maiores chargers Ã© {media_dez_maiores_chargers :.2f}')
```

```python
# Q5

distance1 = chargers.quantile(0.25)
distance2 = chargers.quantile(0.75)
distance_interquartis = distance2 - distance1
print(f'Q5 - A distancia Ã© de {distance_interquartis :.2f}')
```

```python
umemeio_iqr = distance_interquartis * 1.5
limite_inferior = distance1 - umemeio_iqr
limite_superior = distance2 + umemeio_iqr
```

```python
limite_inferior
```

```python
limite_superior
```

```python
chargers.nlargest(15)
```

```python
outliers_superior = df[chargers > limite_superior]
outliers_superior
```

```python
outliers_inferior = df[chargers < limite_inferior]
outliers_inferior
```

```python
outliers_superior.nlargest(20, 'charges')
```
