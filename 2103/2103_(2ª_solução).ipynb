```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
%matplotlib inline
import seaborn as sns
```

#Exercicios

##Exercicio 1

```python
df = sns.load_dataset('fmri')
df
```

```python
df.sort_values(by='timepoint')
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'stim')].sort_values('timepoint').timepoint,
         df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'stim')].sort_values('timepoint').signal,
         color='blue', linestyle='solid', linewidth=3, markersize=1)
plt.title('Subjetc S0')
plt.ylabel('Signal')
plt.xlabel('Time')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'stim')].sort_values('timepoint').timepoint,
     df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'stim')].sort_values('timepoint').signal,
     color='blue', linestyle='dashdot', linewidth=3, markersize=1)
plt.plot(
    df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'cue')].sort_values('timepoint').timepoint,
    df[(df['subject'] == 's0') & (df['region'] == 'parietal') & (df['event'] == 'cue')].sort_values('timepoint').signal,
    color='red', linestyle='dashdot', linewidth=3, markersize=1)
plt.legend(['stim', 'cue'])
plt.title('Subjetc S0')
plt.ylabel('Signal')
plt.xlabel('Time')
plt.show()
```

```python
fig = plt.figure(figsize=(16, 8))
colors = ['red', 'pink', 'green', 'yellow', 'blue', 'black', 'orange', 'brown']


ax = fig.subplots(4, 1)

for i in range(0, 4):
  ax[i].plot(df[(df['subject'] == f's{i}') & (df['region'] == 'parietal') & (df['event'] == 'stim')].sort_values('timepoint').timepoint,
      df[(df['subject'] == f's{i}') & (df['region'] == 'parietal') & (df['event'] == 'stim')].sort_values('timepoint').signal,
      color=colors[i], linestyle='dashdot', linewidth=3, markersize=1)
  ax[i].plot(df[(df['subject'] == f's{i}') & (df['region'] == 'parietal') & (df['event'] == 'cue')].sort_values('timepoint').timepoint,
      df[(df['subject'] == f's{i}') & (df['region'] == 'parietal') & (df['event'] == 'cue')].sort_values('timepoint').signal,
      color=colors[i+4], linestyle='dashdot', linewidth=3, markersize=1)
  ax[i].set_title(f'Subject S{i}')
  ax[i].legend([f'stimS{i}', f'cueS{i}'])
  ax[i].set_xlabel('Time')
  ax[i].set_ylabel('Signal')

plt.show()
```
