```python
# Q5

lista_moedas = ['BTC-USD', 'ITUB4.SA', 'PETR4.SA']
lista_moedas2 = ['BRL=X', 'BTC-USD', 'ITUB4.SA', 'PETR4.SA', '^BVSP']

for c in lista_moedas:
  df_friday[c] = df_friday[c] * df_friday['BRL=X']

df_friday
```

```python
for d in lista_moedas2:
  df_friday[d] = df_friday[d] / df_friday[d].max()

graphic_df_friday = df_friday.reset_index()
graphic_df_friday
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15,4))

df_Friday = graphic_df_friday.set_index('Date')

for c in ['BRL=X', 'BTC-USD', 'ITUB4.SA', 'PETR4.SA', '^BVSP']:
  plt.plot(df_Friday[c], label=c)

plt.title('Friday Adj Close', fontsize=14, weight='bold')
plt.xlabel('Date')
plt.ylabel('Normalized Values')
plt.legend()

plt.show()
```
