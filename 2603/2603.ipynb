#Bibliotecas

```python
import pandas as pd
```

#Arquivos

```python
path = 'https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/data/'

company_before = 'company_before.csv'
company_fev = 'company_fev.csv'
company_mar = 'company_mar.csv'
```

## Primeiro

```python
df_before = pd.read_csv(path + company_before)
df_before
```

## Segundo

```python
df_fev = pd.read_csv(path + company_fev)
df_fev
```

## Terceiro

```python
df_mar = pd.read_csv(path + company_mar)
df_mar
```

#Dropando elementos

## Dropa id do primeiro arquivo

```python
df1 = df_before.drop(columns='id')
df1
```

## Dropa id do segundo arquivo

```python
df2 = df_fev.drop(columns='id')
df2
```

```python
df_mar
```

```python
df3 = df_mar.rename(columns={'Sex':'Sexo', 'Departamento':'Setor', 'Experiencia':'Anos_Experiencia'})
df3
```

```python
df3.Sexo = df3.Sexo.replace('M', 'Masculino')
df3
```

```python
df3.Sexo = df3.Sexo.replace('F', 'Feminino')
df3
```

```python
df3.Setor = df3.Setor.replace('Recursos Humanos', 'RH')
df3
```

```python
df3 = df3[['Setor', 'Sexo', 'Grau_Instrucao', 'Idade', 'Anos_Experiencia', 'Estado', 'Salario']]
df3
```

```python
df_df = pd.concat([df1, df2, df3])
df_df
```

```python
soma_salario = df_df.groupby('Setor').sum()
tabela_salario = soma_salario.drop(columns=['Idade', 'Anos_Experiencia']).reset_index()
tabela_salario
```

```python
tabela_salario.sum()
```

```python
setor_fin = df_df[ df_df.Setor == 'Financeiro' ].drop(columns='Setor').reset_index(drop=True)
setor_fin
```

```python
setor_mkt = df_df[ df_df.Setor == 'Marketing' ].drop(columns='Setor').reset_index(drop=True)
setor_mkt
```

```python
setor_opr = df_df[ df_df.Setor == 'Operacao' ].drop(columns='Setor').reset_index(drop=True)
setor_opr
```

```python
setor_rh = df_df[ df_df.Setor == 'RH' ].drop(columns='Setor').reset_index(drop=True)
setor_rh
```
